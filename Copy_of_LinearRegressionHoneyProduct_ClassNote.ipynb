This file implements Normal Regression to solve kinear regression problem for the honeybee production example.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
from pandas.io.formats.format import set_option
datafile="honeyproduction.csv"
DataReader=open(datafile,mode='rt')
DataRaw=pd.read_csv(DataReader)
DataRaw.head(10)
```

```python
YearData=DataRaw.groupby(['year'])
print(YearData.head())
AnualTotalProduct=YearData.totalprod.sum().reset_index()
print(AnualTotalProduct.head(15))
```

```python
xData=AnualTotalProduct['year']
yData=AnualTotalProduct['totalprod']
plt.scatter(xData,yData)
plt.show
```

```python
x1=xData.values
XT=np.array([np.ones(len(x1)),x1])
X=np.transpose(XT)
phi=np.matmul(XT,X)
phiinv=np.linalg.inv(phi)
theta=np.matmul(np.matmul(phiinv,XT),yData)
#theta=np.dot(np.matmul(phiinv,XT),yData)
print(theta)
yhat=np.matmul(theta,XT)
yError=yData.values-yhat
xbeyond=np.array(range(2013,2021))
Xbeyond=np.array([np.ones(len(xbeyond)), xbeyond])
ybeyond=np.matmul(theta,Xbeyond)
plt.scatter(x1,yData)
plt.scatter(x1,yhat,color='g')
plt.scatter(xbeyond,ybeyond,color='r')
plt.show
```

```python
#import sklearn
from sklearn.linear_model import LinearRegression
model=LinearRegression()
xModel=x1.reshape(-1,1)
model.fit(xModel,yData)
#print(model.coef_)
#print(model.intercept_)
y_hat=model.predict(xModel)
x_beyond=np.array(range(2013,2021))
x_beyond=x_beyond.reshape(-1,1)
y_beyond=model.predict(x_beyond)
plt.scatter(xModel,yData)
plt.scatter(xModel,y_hat,color='g')
plt.scatter(x_beyond,y_beyond,color='r')
plt.show
```
